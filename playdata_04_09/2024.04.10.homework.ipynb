```python
unique_seven = seven_df[~seven_df["store_name"].duplicated()].copy()
```

```python
unique_seven.shape
```

```python
unique_seven.isnull().sum() 
```

```python
unique_seven['sido'] = unique_seven.store_address.apply(lambda x : x.split()[0]) #첫번째값 가져오기 
```

```python
unique_seven.value_counts('sido')
```

```python
service_set=set()
unique_seven['store_service'].apply(lambda x : [service_set.add(y) for y in x])
```

```python
#패턴잡기
unique_seven["24시간"]= unique_seven.store_service.apply(lambda x : True if "24시간" in x else False)
```

```python
unique_seven.value_counts('24시간',normalize=True)
```

```python
for service in service_set:
    # print(x)
    unique_seven[service]= unique_seven.store_service.apply(lambda x : True if service in x else False)
```

```python
unique_seven.drop(['store_service'],axis=1,inplace=True)
```
