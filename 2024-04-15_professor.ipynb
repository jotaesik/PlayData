```python
def fibo(n):
    if n < 2:
        return n
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a
```

```python
%%timeit #시간을 재는것
fibo(10000)
```

```python
def fibo_recu(n):
    if n < 2 : return n
    return fibo_recu(n-1) + fibo_recu(n-2)
```

```python
%%timeit
fibo_recu(10000)
```

```python
def fib_generator():
    a, b = 0, 1 
    while True:
        yield b 
        a, b = b, a + b

fg = fib_generator()
```
