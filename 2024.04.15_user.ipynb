```python
#피보나치수열 def 함수만들기
def func(number):
    # 1 1 2 3 5 8
    # 0 1 2 3 4 5
    # f(5)=5
    if number==1 or number==2:
        return 1
    # elif number==2
    #     return number
    else:
        return func(number-1)+func(number-2)
# func(1)        
```

```python
# 교수님 재귀함수
def fibo_recu(n):
    if n<2 : return n
    return fibo_recu(n-1)+fibo_recu(n-2)

%%timeit #시간계싼
fibo_recu(40)
#코루틴 corountine return을만나면 멈추는게아닌 yield를 마나야 멈춤
```

```python
# 교수님의 for문
from tqdm import tqdm
def fibo(n):
    if n<2:
        return n
    a,b=0,1
    for i in tqdm(range(n)):
        a,b=b,a+b
    return a
# vscode로 디버깅확인해보기
# step over 함수안들어가지만 값은받은 
# step into 함수들어감
# global 전역변수 선언해서 외부에서 확인
# fibo(20577)
```

```python
# yield 
# iterator
def fib_generator():
    a,b=0,1
    while True:
        yield b
        
        a,b = b, a+b
        
```

```python
fg=fib_generator()
```

```python
for _ in range(10): # _ 그냥 의미없다
    print(next(fg)) # yield가 있어서 무한루프가아님 next로 다움 iterator 가져옴
    # stack 은 메모리가 필요하지만 yield는 적당한 메모리안에서 해결가능
```
