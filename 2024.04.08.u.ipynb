```python
import zipfile #zipfile c말고도 java로 된 파이썬 json
import os
import pandas as pd
import shutil 
```

```python
# with구문은 with안에 종속
with zipfile.ZipFile("./이름_생년_성별_10000.zip","r") as f:
    f.extractall("./data")
```

```python
gender=os.listdir("./data")[0].split("_")[-1][0]
```

```python
print(gender)
```

```python
male_list=[]
female_list=list()  #빈리스트 만드는 방법 [] or list()

if not os.path.isdir("./남성"):
    os.mkdir("./남성")
if not os.path.isdir("./여성"): # os.path.isdir("./여성")==False
    os.mkdir("./여성")    

for file in os.listdir("./data"):
    gender=file.split("_")[-1][0]
    # if gender=="1" or gender== "3":
    #     print(gender)
    if gender in ["1","3"]:
        # print("남성",end=" ,")  #end문을 넘어서 print문 통제
        with open(f"./data/{file}","r",encoding="utf-8") as f: #with iopen을 하면 자동으로 close 해주므로 닫아줄필요가없다
            # tmp=f.read()
            male_list.append(['남성',f.read().replace(",","")])
        shutil.move(f"./data/{file}",f"./남성/{file}")
        # with안에 있으면 안되는이유는 잡고있으면 삭제하려고 하는데 어딜삭제라고 에러가남
            # print(tmp)
    else: 
        # print("여성",end=" ,")
        with open(f"./data/{file}","r",encoding="utf-8") as f: 
            # tmp=f.read()
            female_list.append(['여성',f.read().replace(",","")])
        shutil.move(f"./data/{file}",f"./여성/{file}")    
        # pass excpet문에 추가하면 아무것도 하지 않는다
        # ls -l | grep ^- | wc
        # ls -l | grep ^- | wc -l
         # ls -l | grep ^d | wc -l 폴더몇개인지확인
```

```python
male_df=pd.DataFrame(male_list,columns=['성별','금액']) #이중list로 넘길때 colums지정 list안에 dict가있으면 알아서해준
female_df=pd.DataFrame(female_list,columns=['성별','금액'])
```

```python
df=pd.concat([male_df,female_df]) #축을기준이냐 로우기준이냐 기본적으로 세로로한다 
#가로는 axis옵션 merge or join
```

```python
from datetime import date,datetime
# 1997년 04월 03일
```

```python
datetime.strptime("1997년 04월 03일","%Y년 %m월 %d일") #year 4자리면 Y 2자리면 y
# datetime.strptime("97년 04월 03일","%y년 %m월 %d일") #year 4자리면 Y 2자리면 y
```

```python
kbo['생년월일'] = kbo['생년월일'].apply(lambda x : datetime.strptime(x.strip(),"%Y년 %m월 %d일")) #year 4자리면 Y 2자리면 y
```

```python
kbo['생존일'] = kbo['생년월일'].apply(lambda x : (datetime.now()-x).days) #diff
```

```python
kbo.sort_values(by=['생존일',"선수명"],ascending=[False,True]) #기준ㄴ은 생존일->선수명
```
