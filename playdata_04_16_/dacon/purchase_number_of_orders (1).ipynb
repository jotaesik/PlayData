```python
import pickle
import pandas as pd
```

```python
pd.options.display.max_columns = None
```

```python
with open('./dacon_total_df_240504_00.pkl', 'rb') as f:
    df = pickle.load(f)
```

```python
# 날짜 컬럼 추가
df['day'] = df.Order_purchase_timestamp.dt.day
df['YearMonthDay'] = df['Order_purchase_timestamp'].dt.strftime('%Y%m%d')
```

```python
# 90193 rows × 47 columns
df
```

```python
# 90193 rows × 47 columns
df_copy = df.copy()
df_copy
```

```python
# 90193 rows × 48 columns 
# 날짜별 거래건수 합산
df_copy2 = df_copy.groupby(['YearMonthDay', 'order_count','Customer_unique_id']).value_counts().reset_index()
df_copy2
```

```python
# 86467 rows × 4 columns
ddf3 = df_copy2[['YearMonthDay', 'Customer_unique_id', 'order_count']].value_counts().reset_index()
ddf3
```

```python
ddf3['재구매'] = ddf3['order_count'].apply(lambda x : 1 if x > 1 else 0)
```

```python
# 86467 rows × 5 columns
ddf3
```

```python
# 컬럼명 변경
ddf3.rename(columns={'count':'당일구매'},inplace=True)
```

```python
일별_구매건수 = ddf3.groupby(by='Customer_unique_id')['Customer_unique_id'].value_counts().reset_index()
일별_구매건수
```

```python
일별_구매건수[일별_구매건수['Customer_unique_id'] == 'CUSTOMER_Unique_04511']
```

```python
일별_구매건수.rename(columns={'count':'일별구매'},inplace=True)
```

```python
# 84130 rows × 2 columns
일별_구매건수
```

```python
# 86467 rows × 6 columns
ddf4 = ddf3.merge(일별_구매건수, how='inner', on='Customer_unique_id')
ddf4
```

```python
# 180 rows × 6 columns
ddf4[ddf4.일별구매 > 3]
```

```python
with open('./dacon_total_df_240505_00.pkl', 'wb') as f:
    pickle.dump(ddf4, f)
```
